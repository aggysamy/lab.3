```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.countplot(x = 'Response', data=data)
plt.ylabel('Total number of Responses')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Response', hue='Sales Channel', data=data)
plt.ylabel('Responses')
plt.title('Responses by Sales Channel')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(y='Total Claim Amount' , x='Response', data=data)
plt.ylabel('Response by Total Claim Amount')
plt.show()
```
